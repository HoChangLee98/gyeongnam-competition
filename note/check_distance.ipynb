### 시설 거리 데이터
* facility_dist : 시설 별 가까운 정류장 10개 추출
  * 거리의 경우 km 단위
* facility_location : 시설 별 가까운 정류장 10개에 대한 정보 데이터 저장
  * pd.DataFrame(facility_location["시설명"])

```python
import pandas as pd
import pickle

with open("../data/거리계산/facility_dist_result_df.pkl", "rb") as f:
    facility_dist = pickle.load(f)

with open("../data/거리계산/facility_location_results.pkl", "rb") as f:
    facility_location = pickle.load(f)        

import geopandas as gpd
```

```python
## 시설 위치
facility = pd.read_csv(
    filepath_or_buffer="../data/시설.csv"
)
facility = facility.dropna()
facility = gpd.GeoDataFrame(
    facility, 
    geometry=gpd.points_from_xy(facility['경도'], facility['위도'])
    ).set_crs(crs="EPSG:4326").to_crs(epsg=5179)
facility = facility.reset_index(drop=True)
```

```python
facility[facility["시설명"] == "서촌"]
```

```python
list(facility["시설명"])
```

```python
facility_dist.sort_values(by="정류장1", ascending=False).head(50)
```

```python
pd.DataFrame(facility_location["덕산한의원"])
```
