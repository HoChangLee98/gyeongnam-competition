```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import geopandas as gpd
import glob
import os
from shapely.geometry import Polygon
from shapely.geometry import Point
```

### 노인 인구 밀집 지역- 거제

```python
gdf = gpd.read_file(
    "../data/경남 노인 인구 분포/거제시/nlsp_021001010.shp", 
    encoding="utf-8", 
    )
gdf['val'] = gdf['val'].fillna(0)
gdf = gdf[gdf["val"].apply(lambda x: x != 0.0)]
gdf = gdf.to_crs(epsg=5179)
```

```python
gdf['point'] = gdf['geometry'].centroid.apply(lambda centroid: Point(centroid.x, centroid.y))
```
