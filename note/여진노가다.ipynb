```python
# 빈 딕셔너리 초기화
result_dict = {}

# 데이터에서 nodeid와 routeid 추출하여 딕셔너리 형태로 추가
for i in data['38010'].keys():
    for j in range(len(data['38010'][i])):
        nodeid = data['38010'][i][j]['nodeid']
        routeid = data['38010'][i][j]['routeid']
        
        if nodeid not in result_dict:
            result_dict[nodeid] = {}
        
        if routeid in result_dict[nodeid]:
            # 이미 해당 nodeid와 routeid에 대한 데이터가 존재할 경우, 리스트로 변환하여 데이터 추가
            if isinstance(result_dict[nodeid][routeid], list):
                result_dict[nodeid][routeid].append(data['38010'][i][j])
            else:
                result_dict[nodeid][routeid] = [result_dict[nodeid][routeid], data['38010'][i][j]]
        else:
            # 새로운 routeid와 해당 데이터를 추가합니다.
            result_dict[nodeid][routeid] = data['38010'][i][j]

# 결과 딕셔너리 출력
print(result_dict)
```
