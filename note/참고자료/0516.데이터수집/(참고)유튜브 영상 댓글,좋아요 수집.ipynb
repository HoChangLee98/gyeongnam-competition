### URL = https://www.youtube.com/watch?v=igw325ntuAI

#### 수집 가능한 정보
> 댓글 아이디, 댓글 업데이트 날짜, 댓글 내용, 좋아요 수

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
driver = wb.Chrome()
driver.get("https://www.youtube.com/watch?v=igw325ntuAI")
```

```python
for i in range(1,30) :
    body = driver.find_element(By.CSS_SELECTOR, "body")
    body.send_keys(Keys.END)
    time.sleep(1)
```

```python
review = driver.find_elements(By.CSS_SELECTOR, "#content-text")
like = driver.find_elements(By.CSS_SELECTOR, "#vote-count-middle")
```
