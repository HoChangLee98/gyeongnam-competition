```python
driver = wb.Chrome()
time.sleep(0.5)
driver.get("https://www.youtube.com/@Busan_is_good/videos")
time.sleep(1)
body = driver.find_element(By.CSS_SELECTOR, "body")
for i in range(10) :
    body.send_keys(Keys.END)
    time.sleep(1)

# 영상 제목    
title = driver.find_elements(By.ID, "video-title")
#영상 조회수
view = driver.find_elements(By.CSS_SELECTOR, "#metadata-line > span:nth-child(3)")
#영상 링크
url = driver.find_elements(By.ID, 'video-title-link')

titleList = [] # 영상의 제목들이 담을 리스트
viewList = [] # 영상의 조회수들이 담을 리스트
urlList = [] # 영상의 링크들이 담을 리스트

#수집된 내용들을 하나씩 꺼내서 각각의 리스트에 담음
for i in range(len(title)) :
    titleList.append(title[i].text)
    viewList.append(view[i].text)
    urlList.append(url[i].get_attribute('href'))
print(len(titleList))
print(titleList)
driver.quit()
```

```python
titleList[0]
```

```python
viewList[0]
```

```python
urlList
```

```python
import pandas as pd 

yutube_data = {'Title': titleList,
               'URL' : urlList,
               'Views': viewList}
yutube_df = pd.DataFrame(yutube_data)
yutube_df
```

### [실습] 최종 부산튜브 업로드 동영상 데이터프레임 만들기
#### 위의 데이터프레임에 유튜브 영상 업데이트 날짜(updated_date) 컬럼 추가하기

```python
driver = wb.Chrome()
time.sleep(0.5)
driver.get("https://www.youtube.com/@Busan_is_good/videos")
time.sleep(1)
body = driver.find_element(By.CSS_SELECTOR, "body")
for i in range(10) :
    body.send_keys(Keys.END)
    time.sleep(1)

dates = driver.find_elements(By.CSS_SELECTOR, "#metadata-line > span:nth-child(4)")
dateList = []
for d in dates:
    dateList.append(d.text)
    
print(len(dateList))
driver.quit()
```

```python
yutube_df["updated_date"] = dateList
```
