```python
import folium
from folium.plugins import MarkerCluster
m = folium.Map(location = [data['위도'][0],data['경도'][0]], zoom_start=13, tiles='cartodbpositron')

folium.GeoJson(
    sgg,
    name = 'SGG_NM'
).add_to(m)

mc = MarkerCluster()

for _, row in data.iterrows() :
    mc.add_child(
        folium.Circle(location=(row['위도'], row['경도']), popup=row['장기요양기관'], radius=1000, color='#FF580B',
            fill='#FF580B')
    )
    mc.add_child(
        folium.Marker(location=(row['위도'], row['경도']), popup=row['장기요양기관'], color='#FF580B',
                fill='#FF580B')
    )
m.add_child(mc) 
m #
```
